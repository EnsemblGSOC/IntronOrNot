```python
chrom_uniques = list(df_repeat_features.Chrom.unique())
df_chrom = {}

# Splitting the dataframe rows separately
for chrom in tqdm(chrom_uniques):
    df_chrom[chrom] = df_repeat_features[df_repeat_features.Chrom == chrom]

df_chrom
```
