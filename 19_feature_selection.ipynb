```python
from sklearn.feature_selection import RFECV
import pandas as pd
from xgboost import XGBClassifier
from sklearn import preprocessing
import sklearn
```

```python
X = pd.read_csv("data/17_trainset_featurised.tsv", sep="\t")
y = X.pop("class")
```

```python
le = preprocessing.LabelEncoder()
le.fit(X["splice_site"])
X["splice_site"] = le.transform(X["splice_site"])

le = preprocessing.LabelEncoder()
le.fit(X["chr"])
X["chr"] = le.transform(X["chr"])

le = preprocessing.LabelEncoder()
le.fit(X["bps"])
X["bps"] = le.transform(X["bps"])

X = X.drop(["strand"], axis=1)
```

```python
X = X.fillna(0)
```

```python
%%time

estimator = XGBClassifier(n_jobs=20)
selector = RFECV(estimator, min_features_to_select=1, step=1, verbose=1, cv=5, scoring="matthews_corrcoef")
selector = selector.fit(X, y)
```

```python
feature_lst = X.columns.tolist()
```

```python
feature_rank_name_pairs = list(zip(feature_lst, selector.ranking_))
feature_rank_name_pairs
```

```python
feature_rank_name_pairs = sorted(feature_rank_name_pairs, key=lambda x: x[1])
feature_rank_name_pairs
```
