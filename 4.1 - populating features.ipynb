```python
import pandas as pd
import ast
```

```python
df = pd.read_csv("data/3_all_features.tsv", sep="\t")
# df.drop(columns="intron_id", inplace=True)
df
```

```python
df['RC3'] = df['RC3'].replace("[None, -1]", "[None -1]")
df['RC3'] = df['RC3'].str.replace('[\[\]]', '')
df['RC3'] = df['RC3'].str.replace('\n', '')
df
```

```python
df["RC3"].str.split(" ").str.get(0).str.replace("'", "").str.split(";")
```

```python
df_featuers_split = df["RC3"].str.split(" ").str.get(0).str.replace("'", "").str.split(";")
df["GTExv2"] = df_featuers_split.str[0].str.split("=").str[1]
df["TCGAv2"] = df_featuers_split.str[1].str.split("=").str[1]
df["SRAv3h"] = df_featuers_split.str[2].str.split("=").str[1]
df["RC3-Splice_site"] = df_featuers_split.str[3]
df["RC3-Score"] = df["RC3"].str.split(" ").str.get(1)
```

```python
df
```

```python
df = df.drop(columns="RC3")
df
```

# Populate the repeat_features into one-hots

```python
df['lst_repeat_features_start_site'] = df['repeat_features_start_site'].apply(lambda x: ast.literal_eval(x) if len(x) > 2 else [])
df['lst_repeat_features_end_site'] = df['repeat_features_end_site'].apply(lambda x: ast.literal_eval(x) if len(x) > 2 else [])
df
```

```python
unique_repeat_features_values_start_site = set()
unique_repeat_features_values_end_site = set()
for idx, row in df.iterrows():
    unique_repeat_features_values_start_site.update(row['lst_repeat_features_start_site'])
    unique_repeat_features_values_end_site.update(row['lst_repeat_features_end_site'])

unique_repeat_features_values_start_site
```

```python
unique_repeat_features_values_end_site
```

```python
for repeat_feature in unique_repeat_features_values_start_site:
    df["repeat_features_start_site:"+repeat_feature] = df["repeat_features_start_site"].str.count("'"+repeat_feature+"'").astype(int)


for repeat_feature in unique_repeat_features_values_start_site:
    df["repeat_features_end_site:"+repeat_feature] = df["repeat_features_end_site"].str.count("'"+repeat_feature+"'").astype(int)
df
```

# Now we drop all redundant columns

```python
df = df.drop(columns=["repeat_features_start_site", "repeat_features_end_site", "lst_repeat_features_start_site", "lst_repeat_features_end_site"])
df
```

```python
df[df["class"] == 0]
```

```python
df_splice_site_split = df["RC3-Splice_site"].str.split(":")
df["RC3 - Acceptor splice site"] = df_splice_site_split.str[0]
df["RC3 - Donor splice site"] = df_splice_site_split.str[1]
df
```

```python
df.to_csv("data/4_populated_features.tsv", sep="\t", index=False)
```
